# Lenguaje Python
* En la celda de abajo buscamos en la columna de Programming Language todos los que contenga la palabra Python
de esta forma para poder identificar cuantos artículos usar python. Además, se muestran los autores de correspondencia.
* Los resultado se compararon con la tabla de excely muestra los mismos resultados para el lenguaje python.

```python
# Esto busca la palabra Python en la columna de Programming Language, y nos da True y False cuando
# encuentra la palabra y lo guarda en la variable Py.
Py=df['7'].str.contains('Python')== True

#esto nos da el data frame con los autores en donde esta la palabra python y lo asigna a DF_Py
DF_Py=df[df['7'].str.contains('Python')== True].loc[:,['1','7']]
          
# esto nos da cuantos valores son para NaN, True y False para la palabra python y lo asigna a Py_Srs    
Py_Srs=pd.Series(df['7']).str.contains('Python',regex=False).value_counts(dropna=False)    
    
# esto imprime los valores True y False para la palabra Python.
print(Py.value_counts(),Py_Srs,DF_Py.shape,sep='\n\n')
DF_Py.reset_index() # Python dataframe
```
