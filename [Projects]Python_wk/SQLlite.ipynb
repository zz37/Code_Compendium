```python
# para realizar consultas debemos establecer un cursor
cursorObj = conn.cursor() 
# un cursor de una base de datos es una estructura de control que sirve para trasvesar sobre
# los registros de la base de datos

# una vez que tenemos el cursor podemos usar el objeto de cursos para ejecutar consultas
cursorObj.execute('SELECT player_name from Player') 

# cursorObj.fetchall() también genera el mismo resultado
for row in cursorObj.fetchall():
    print(row)  
```

```python
# aquí seleccionamos la altura de los jugadores

cursorObj = conn.cursor()
cursorObj.execute('SELECT height from Player')
for row in cursorObj.fetchall():
    print(row)
```

```python
# sucede lo mismo pero 
cursorObj.execute('SELECT height from Player')
# pero aquí en este for aplanamos la lista eligiendo el índice 0 de todas las tuplas y lo guardamos en una lista, con
# el método de list comprehension
player_heiht = [row[0] for row in cursorObj.fetchall()]
print(min(player_heiht))
print(max(player_heiht))
```

```python
# importamos el numpy y convertimos la lista en un np array
import numpy as np
player_heiht = np.array(player_heiht)
```

```python
# obtenemos métricas de la altura de los jugadores
print(player_heiht.mean())
print(player_heiht.min())
print(player_heiht.max())
print(np.quantile(player_heiht,q = [0.25, 0.50, 0.75]))
```

```python
# en este blocque seleccionamos la altura y el peso e imprimimos solo los primeros 5 jugadores
cursorObj.execute('SELECT height, weight from Player')
players = [row for row in cursorObj.fetchall()]
for n in range(5):
    print(players[n])
```

```python
players = np.array(players) # convertimos a un np array
```

```python
# obtenemos métricas de la altura de los jugadores
print("Promedio ",players.mean(axis=0))
print("Minimo ", players.min(axis=0))
print("Maximo ", players.max(axis=0))
print("Cuartiles ", np.quantile(players,q = [0.25, 0.5, 0.75],axis=0))
```

## SQL con Pandas

```python
import sqlite3
db_file = 'data/soccer/database.sqlite'
conn = sqlite3.connect(db_file)
```

```python
# guardamos la consulta en una variable
q = '''SELECT League.name,Match.home_team_goal,Match.away_team_goal,Match.date 
                  FROM League INNER JOIN Match ON Match.league_id==League.id '''
```

```python
# pasamos a pandas la consulta y la conexión de la base de datos
import pandas as pd
df =  pd.read_sql(q,conn)
```

```python
df.head() # vemos la tabla
```

```python
df['diff'] = (df['home_team_goal']-df['away_team_goal']).abs() # podemos crear una nueva columna de la dif. de goles
```

```python
df.head() #mostrando
```

```python
df.sort_values(by='diff',ascending=False).head() # sorteando de mayor a menor
```

```python
# con seaborn podemos ver la distribución de los goles de casa vs goles de visitante
import seaborn as sns
sns.set(style="darkgrid")
sns.jointplot(x="home_team_goal",y="away_team_goal",kind="kde", data=df)
```
